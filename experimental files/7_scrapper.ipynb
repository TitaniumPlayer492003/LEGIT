```python
import time
import random
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

# Manually enter the case URL
case_url = "https://indiankanoon.org/doc/173448784/"

# Extract case ID from URL
case_id = case_url.rstrip("/").split("/")[-1]
filename = f"{case_id}.txt"

# Chrome options to evade bot detection
options = uc.ChromeOptions()
options.add_argument(
    "--headless"
)  # Run in background (remove if you want to see the browser)
options.add_argument(
    "--disable-blink-features=AutomationControlled"
)  # Hide bot detection
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.6834.159 Safari/537.36"
)  # Spoof a real browser

# Start Undetected ChromeDriver
driver = uc.Chrome(options=options)

# Open the case URL
driver.get(case_url)

# **Human-like behavior**
time.sleep(random.uniform(5, 10))  # Random wait time

# Scroll down slightly
driver.execute_script("window.scrollBy(0, 300);")
time.sleep(random.uniform(2, 5))

# Move the mouse randomly
actions = ActionChains(driver)
actions.move_by_offset(random.randint(10, 100), random.randint(10, 100)).perform()
time.sleep(random.uniform(1, 3))

try:
    title = driver.find_element(By.TAG_NAME, "h1").text  # Case title
    date = driver.find_element(By.CLASS_NAME, "docdate").text  # Case date
    case_number = driver.find_element(By.CLASS_NAME, "casenumber").text  # Case number
    content = driver.find_element(By.CLASS_NAME, "judgments").text  # Full case text

    # Save data to a text file
    with open(filename, "w", encoding="utf-8") as file:
        file.write(f"Title: {title}\n")
        file.write(f"Date: {date}\n")
        file.write(f"Case Number: {case_number}\n")
        file.write("\n--- Case Content ---\n")
        file.write(content)

    print(f"✅ Data saved to {filename}")

except Exception as e:
    print(f"❌ Error extracting data: {e}")

# Close browser
driver.quit()
```
