```python
import time
import random
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Input case URL manually
case_url = "https://indiankanoon.org/doc/173448784/"  # Change this for each case

# Extract case ID from URL (last n characters before "/")
case_id = case_url.rstrip("/").split("/")[-1]  # "173448784"
filename = f"{case_id}.txt"

# Set up Chrome options to bypass Cloudflare bot detection
options = Options()
options.add_argument("start-maximized")
options.add_argument("disable-infobars")
options.add_argument("disable-extensions")
options.add_argument("disable-gpu")
options.add_argument("headless")  # Optional: run browser without GUI
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
)  # Replace with a real browser user-agent

# Initialize WebDriver
driver = webdriver.Chrome(options=options)

# Open the case URL
driver.get(case_url)

# Random sleep to simulate human behavior
time.sleep(random.uniform(5, 10))  # Random delay to avoid looking like a bot

# Wait for the page to load completely using WebDriverWait
try:
    # Wait for the title, case date, case number, and content to be present
    title = (
        WebDriverWait(driver, 10)
        .until(EC.presence_of_element_located((By.TAG_NAME, "h1")))
        .text
    )

    date = (
        WebDriverWait(driver, 10)
        .until(EC.presence_of_element_located((By.CLASS_NAME, "docdate")))
        .text
    )

    case_number = (
        WebDriverWait(driver, 10)
        .until(EC.presence_of_element_located((By.CLASS_NAME, "casenumber")))
        .text
    )

    content = (
        WebDriverWait(driver, 10)
        .until(EC.presence_of_element_located((By.CLASS_NAME, "judgments")))
        .text
    )

    # Save to a text file
    with open(filename, "w", encoding="utf-8") as file:
        file.write(f"Title: {title}\n")
        file.write(f"Date: {date}\n")
        file.write(f"Case Number: {case_number}\n")
        file.write("\n--- Case Content ---\n")
        file.write(content)

    print(f"✅ Data saved to {filename}")

except Exception as e:
    print(f"❌ Error extracting data: {e}")

# Close the browser
driver.quit()
```
